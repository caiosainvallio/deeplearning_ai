```python
from langchain.chains.combine_documents.stuff import StuffDocumentsChain
from langchain.chains import MapReduceDocumentsChain, ReduceDocumentsChain
from langchain.text_splitter import CharacterTextSplitter
from langchain.chains.llm import LLMChain
from langchain.prompts import PromptTemplate

llm = ChatOpenAI(temperature=0, model='gpt-4-1106-preview')

# Map
map_template = """A seguir está um conjunto de documentos contendo trechos de um talk show
{docs}
Com base nesta lista de documentos, descreva o trecho de forma resumida. 
Resposta útil:"""
map_prompt = PromptTemplate.from_template(map_template)
map_chain = LLMChain(llm=llm, prompt=map_prompt)
```

```python
# Reduce
reduce_template = """A seguir está um conjunto de resumos de trechos de um talk show:
{doc_summaries}
Pegue-os e transforme-os em um texto texto de modo irreverente e coloquial, em primeira pessoa, com o apresentador narrando como foi a entrevista. 
O nome do talk show é "Achismos" e o nome do apresentador é Maurício Meirelles.
Garanta que o texto final seja coeso e coerente.
Não invente nada, apenas narre como foi a entrevista.
O texto deve ter entres 2500 e 3000 palavras.
Resposta útil:"""
reduce_prompt = PromptTemplate.from_template(reduce_template)
```

```python
# Run chain
reduce_chain = LLMChain(llm=llm, prompt=reduce_prompt)
```

```python
 # Takes a list of documents, combines them into a single string, and passes this to an LLMChain
combine_documents_chain = StuffDocumentsChain(
    llm_chain=reduce_chain, document_variable_name="doc_summaries"
)
```

```python
# Combines and iteravely reduces the mapped documents
reduce_documents_chain = ReduceDocumentsChain(
    # This is final chain that is called.
    combine_documents_chain=combine_documents_chain,
    # If documents exceed context for `StuffDocumentsChain`
    collapse_documents_chain=combine_documents_chain,
    # The maximum number of tokens to group documents into.
    token_max=4000,
)
```

```python
# Combining documents by mapping a chain over them, then combining results
map_reduce_chain = MapReduceDocumentsChain(
    # Map chain
    llm_chain=map_chain,
    # Reduce chain
    reduce_documents_chain=reduce_documents_chain,
    # The variable name in the llm_chain to put the documents in
    document_variable_name="docs",
    # Return the results of the map steps in the output
    return_intermediate_steps=False,
)

text_splitter = CharacterTextSplitter.from_tiktoken_encoder(
    chunk_size=1000, chunk_overlap=0
)
split_docs = text_splitter.split_documents(texts)
```
